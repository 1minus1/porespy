```python
import porespy as ps
import scipy as sp
import matplotlib.pyplot as plt
```

```python
im = ps.generators.cylinders(shape=[200, 200, 200], radius=5, ncylinders=400, phi_max=90, theta_max=0)
```

```python
fig, ax = plt.subplots(1, 3, figsize=[15, 15])
ax[0].imshow(im[100, :, :])
ax[1].imshow(im[:, 100, :])
ax[2].imshow(im[:, :, 100])
```
